```python
def generate_random_colors(n):
    return ['#%02X%02X%02X' % tuple(np.random.randint(0, 256, 3)) for _ in range(n)]
```

```python
results = [[model, np.random.randint(1, 11)] for model in models]
df = pd.DataFrame(results, columns=["Model", "Score"])
colors = generate_random_colors(len(models))

plt.figure(figsize=(14, 8))
plt.barh(df["Model"], df["Score"], color=colors)

plt.xlabel("Puntuaciones")
plt.ylabel("Modelos")
plt.title("Puntuaciones Modelos")
plt.xlim(0, 10)
plt.tight_layout()

plt.show()
```
