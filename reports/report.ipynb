```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
runs = [run for run in os.listdir() if run.endswith(".csv") and run != "summary.csv"]
runs = [(run, pd.read_csv(run)) for run in runs]
```

```python
STEPS_FACTOR = 250
```

```python
rows = []
for name, run in runs:
    for col in run.columns:
        if col.endswith(" - eval/loss"):
            rows.append(run[col].rename(f"{col.replace(' - eval/loss', '')} - ({name.removesuffix('.csv')})"))
```

```python
df = pd.DataFrame(rows).transpose()
```

```python
x = [i * STEPS_FACTOR for i in range(1, len(df) + 1)]
fig, ax = plt.subplots()
ax.plot(x, df, label=df.columns)
ax.axhspan(3.2, 8.6, color="blue", alpha=0.3, label="Grupo 1 (3.2, 8.6)")
ax.axhspan(1, 3.2, color="green", alpha=0.3, label="Grupo 2 (1, 3.2)")
ax.axhspan(0.2, 1, color="red", alpha=0.3, label="Grupo 3 (0.2, 1)")
ax.axhspan(-0.1, 0.2, color="yellow", alpha=0.3, label="Grupo 4 (-0.01, 0.2)")
plt.xlabel("Steps")
plt.ylabel("Pérdida")
plt.title("Curvas de pérdida")
plt.xlim(0, 18000)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc="center left", ncol=1)
plt.show()
```

```python
info = df.describe().transpose()
info.sort_values(by="min", ascending=False)
```

```python
g1 = info[info["min"] >= 3.2].sort_values(by="min", ascending=False)
g1
```

```python
g2 = info[(info["min"] <= 3.2) & (info["min"] >= 1)].sort_values(by="min", ascending=False)
g2
```

```python
g3 = info[(info["min"] <= 1) & (info["min"] >= 4e-4)].sort_values(by="min", ascending=False)
g3
```

```python
g4 = info[info["min"] < 4e-4].sort_values(by="min", ascending=False)
g4
```

```python
def plot_data(data, group, lim=None):
    x = [i * STEPS_FACTOR for i in range(1, len(data) + 1)]
    plt.plot(x, data, label=data.columns)
    plt.xlabel("Steps")
    plt.ylabel("Pérdida")
    plt.title(f"Curvas de pérdida ({group})")
    if lim is not None:
        plt.xlim(0, lim)
    plt.legend(bbox_to_anchor=(1.05, 0.5), loc="center left", ncol=1)
    plt.show()
```

```python
plot_data(df[g1.T.columns], "Group 1", 18000)
```

```python
plot_data(df[g2.T.columns], "Group 2")
```

```python
plot_data(df[g3.T.columns], "Group 3", 10000)
```

```python
plot_data(df[g4.T.columns], "Group 4", 3000)
```
